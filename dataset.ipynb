## Create dataset of periodic functions
Creates multiple datasets for multiple functions with variying density

```python
%load_ext autoreload
%autoreload 1
import numpy as np
import pandas as pd
%aimport plotting_functions
import plotting_functions as pf
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
range_multiplier = 3
pi = math.pi
density = [0.001, 0.005, 0.01, 0.02, 0.05, 0.1, 0.2, 0.5]
```

```python
lower_bound = -1*range_multiplier*pi
upper_bound = range_multiplier*pi
```

```python
x_data = []
y_data = []
for idx, each_density in enumerate(density):
    x_data.append(np.arange(lower_bound, upper_bound, each_density))
    y_data.append(np.sin(x_data[idx]))
```

```python
fig, ax = plt.subplots(1, 1)
pf.simple_2D(ax, x_data[0], y_data[0])
```
